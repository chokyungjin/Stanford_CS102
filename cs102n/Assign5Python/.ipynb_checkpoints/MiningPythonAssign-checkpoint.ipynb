```python
# Read movies dataset from CSV file
# Create dictionary "items" with key = item and value = list of transactions
# Also set variable numtrans = number of transactions
items = {}
trans = []  # list of transactions used to set numtrans

with open('Movies.csv','r') as f:
    rows = csv.DictReader(f)
    for r in rows:
        if r['item'] not in items:
            items[r['item']] = [r['TID']]
        else:
            items[r['item']].append(r['TID'])
        if r['TID'] not in trans:
            trans.append(r['TID'])
numtrans = len(trans)
print ('Number of transactions (users):', numtrans)
print ('Number of distinct items (movies):', len(items))

#print ('Item dictionary:', items)

#items['The Fault in Our Stars']
```

#### Use the brute-force (i.e., not Apriori) method to mine for frequent item-sets of three and four items in the Movies dataset. Find a single support threshold where the number of frequent item-sets of three items is more than 10 but less than 20, and the number of frequent item-sets of four items is more than 0. WATCH OUT! The calculation for item-sets of four can be quite slow.

```python

# Frequent item-sets of three
# support = INSERT VALUE HERE
# YOUR CODE HERE

list1 = []

#support = .3

for i1 in items:
    for i2 in items:
        for i3 in items:
            if i1 < i2 and i2 < i3:
#                 print('i1=',i1,set(items[i1]))
#                 print()
#                 print('i2=',i2,set(items[i2]))
#                 print()
#                 print('i3=',i3,set(items[i3]))
#                 print()
                common = len(set(items[i1]) & set(items[i2]) & set(items[i3]))
                
                #print(float(common)/numtrans)
                #print(float(common) / support)
                if float(common) >= 10 and float(common) <= 20:
                    #print (i1, '|', i2, '|', i3)
                    list1.append([i1, '|', i2, '|', i3])

                    
#list1 = set(list1)
print(list1)
```

```python
# Frequent item-sets of four
#support = INSERT VALUE HERE (SAME VALUE AS IN PREVIOUS CELL)
#YOUR CODE HERE

list2 = []

#support = .4

for i1 in items:
    for i2 in items:
        for i3 in items:
            for i4 in items:
                if i1 < i2 and i2 < i3 and i3 < i4:
                    common = len(set(items[i1]) & set(items[i2]) & set(items[i3]) & set(items[i4]))                
                #print(float(common)/numtrans)
                #print(float(common) / support)
                    if float(common) >= 10 and float(common) <= 20:
                        print (i1, '|', i2, '|', i3, '|', i4)
                        list2.append([i1, '|', i2, '|', i3 ,'|' , i4])

                    
#list1 = set(list1)
print(list2)
```

#### EXTRA CREDIT: compute frequent item-sets of four using the Apriori method. Notice how incredibly much faster it is.

```python
# Frequent item-sets of four using Apriori
support = 0.03

#YOUR CODE HERE

ones = []
for i1 in items:
    for i2 in items:
        for i3 in items:
            for i4 in items:
                common = len(set(items[i1]) & set(items[i2]) & set(items[i3]) & set(items[i4]))
                #print(float(common)/numtrans , support)
                if float(common)/numtrans > support:
                    ones.append([i1,i2,i3,i4])
                    #print(ones)

print(ones)
```
